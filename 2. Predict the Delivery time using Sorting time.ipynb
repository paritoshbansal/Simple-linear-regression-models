## Objective  - Is to predict the Delivery time using Sorting time

```python
y=data["Delivery Time"].values.reshape(-1,1)
x=data["Sorting Time"].values.reshape(-1,1)
```

## linear regression model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin=LinearRegression()
```

```python
model=lin.fit(x_train,y_train)
print("training completed")
```

```python
model.score(x_train,y_train)
print("score of training data=",model.score(x_train,y_train))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,model.predict(x_test))
```

```python
print("r2_score of testing data=",r2_score(y_test,model.predict(x_test)))
```

```python
print("r2_score is closer to 1 that means model is strong model")
```
